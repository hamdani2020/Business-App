```python
# Print title
print("*" * 50)
print("Business App".center(50))
print("*" * 50)

import math
class Business:
    def __init__(self, name, easting, northing, details):
        self.name = name
        self.easting = easting
        self.northing = northing
        self.details = details
businesses = []

def add_business():
    name = input("Enter the name of the business: ")
    easting = float(input("Enter the Easting: "))
    northing = float(input("Enter the Northing: "))
    details = input("Enter details about the business: ")
    business = Business(name, easting, northing, details)
    businesses.append(business)
    print(f"Business '{name}' added successfully.")

def print_all_businesses():
    print("All Businesses:")
    for business in businesses:
        print(f"{business.name} - Easting: {business.easting}, Northing: {business.northing}, Details: {business.details}")

def search_by_name(name):
    matching_businesses = [business for business in businesses if business.name.lower() == name.lower()]
    if matching_businesses:
        print(f"Found {len(matching_businesses)} business{'es' if len(matching_businesses) > 1 else ''} with the name '{name}':")
        for business in matching_businesses:
            print(f"{business.name} - Easting: {business.easting}, Northing: {business.northing}, Details: {business.details}")
    else:
        print(f"No business found with the name '{name}'.")

def calculate_distance(easting1, northing1, easting2, northing2):
    return math.sqrt((easting2 - easting1) ** 2 + (northing2 - northing1) ** 2)

def search_by_distance(easting, northing, distance):
    print(f"Businesses within {distance} units of the location (Easting: {easting}, Northing: {northing}):")
    for business in businesses:
        dist = calculate_distance(easting, northing, business.easting, business.northing)
        if dist <= distance:
            print(f"{business.name} - Easting: {business.easting}, Northing: {business.northing}, Details: {business.details}")

def main():
    while True:
        print("\nWelcome to the Business Application!")
        print("1. Add Business")
        print("2. Print All Businesses")
        print("3. Search by Name")
        print("4. Search by Distance")
        print("5. Exit")

        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            add_business()

        elif choice == '2':
            print_all_businesses()

        elif choice == '3':
            name = input("Enter the name to search for: ")
            search_by_name(name)

        elif choice == '4':
            easting = float(input("Enter the Easting of the location: "))
            northing = float(input("Enter the Northing of the location: "))
            distance = float(input("Enter the maximum distance to search: "))
            search_by_distance(easting, northing, distance)

        elif choice == '5':
            print("Thank you for using the Business . Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()

print("*" * 50)
print("Created LusiTech I.T Consults")

```
